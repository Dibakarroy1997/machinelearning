# Random Forest Regression

### Data Processing

```r
# Importing the dataset
dataset = read.csv('Position_Salaries.csv')
dataset = dataset[2:3]
```

### Fitting the Random Forest Regression Model to the dataset

```r
# install.package('randomForest')
library(randomForest)
set.seed(42)
regressor = randomForest(x = dataset[1], 
                         y = dataset$Salary,
                         ntree = 110)
# dataset[] gives a dataframe while dataset$DependentVariable will give us a vector
```

### Visualising the Random Forest Regression Model results (for higher resolution and smoother curve)

```r
# install.packages('ggplot2')
library(ggplot2)
x_grid = seq(min(dataset$Level), max(dataset$Level), 0.01)
ggplot() +
  geom_point(aes(x = dataset$Level, y = dataset$Salary),
             colour = 'red') +
  geom_line(aes(x = x_grid, y = predict(regressor, newdata = data.frame(Level = x_grid))),
            colour = 'blue') +
  ggtitle('Truth or Bluff (Random Forest Regression Model)') +
  xlab('Level') +
  ylab('Salary')
```

### Predicting a new result

```r
y_pred = predict(regressor, data.frame(Level = 6.5))
y_pred
```

* * *

Here the predicted value is almost equal to 160K which is the Salary value prposed by the employee. Also the model seems to be much better then the Polynomial Regression Model.
